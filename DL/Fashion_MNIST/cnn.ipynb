Convolutional neural network (CNN)
Use MNIST Fashion Dataset and create a classifier to classify fashion clothing into categories.

```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import  Conv2D, MaxPooling2D,  Flatten, Dense 
```

```python
(train_x, train_y), (test_x, test_y) = fashion_mnist.load_data()
```

```python
model = Sequential()

model.add(Conv2D(64,(3,3),activation='relu',input_shape=(28, 28, 1)))

# This creates a 2D convolutional layer with several parameters:
# 64: The number of filters/kernels in the convolutional layer. 
# - This controls the dimensionality of the output space.
# (3, 3): The size of the convolutional kernel/filter. 
# - A (3, 3) kernel means a 3x3 matrix is slid across the input image to perform convolutions.
# activation='relu': This specifies the activation function used after the convolution operation. 
# input_shape=(28, 28, 1): This specifies the expected shape of the input to the model. 
# - In this case, the input is an image of size 28x28 with a single channel (like a grayscale image).

# Adding maxpooling layer to get max value within a matrix
model.add(MaxPooling2D())

model.add(Flatten())
model.add(Dense(128, activation = "relu"))
model.add(Dense(10, activation = "softmax"))

'''
model.add(MaxPooling2D(pool_size=(2,2))):
This layer is a max-pooling layer, typically used after a convolutional layer to reduce the spatial 
dimensions of the input. Here's what each component does:
- MaxPooling2D: The type of pooling layer, which takes the maximum value within the specified pool size.
- pool_size=(2,2): This indicates the size of the pooling window. In this case, a 2x2 window is used.

model.add(Flatten()): This layer flattens the output from the previous layer, converting it into a 
single-dimensional array. Flattening is often required to transition from convolutional/pooling layers 
to fully connected (dense) layers.
'''
```

```python
model.summary()
```

```python
model.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['accuracy'])
```

```python
model.fit(train_x.astype(np.float32), train_y.astype(np.float32), epochs = 5, validation_split = 0.2)
```

```python
loss, acc = model.evaluate(test_x, test_y)
```

```python
labels = ['t_shirt', 'trouser', 'pullover', 'dress', 'coat', 'sandal', 'shirt', 'sneaker',   'bag', 'ankle_boots']
```

```python

predictions = model.predict(test_x)
```

```python
import matplotlib.pyplot as plt

for i, prediction in enumerate(predictions):
    
    label_index = np.argmax(prediction)   
    label = labels[label_index]
    
    # Print the prediction result
    print(f"Label: {label}")

    # Plot the corresponding image from test_x
    plt.imshow(test_x[i])
    plt.show()
```
